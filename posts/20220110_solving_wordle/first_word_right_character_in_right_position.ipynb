```python
#read data
f = open('words.json')
data = json.load(f)
f.close()

#we are interested in the distribution of the solution words
solution_words = data['Aa']
useable_words = data['Aa']
```

```python
#3. guess so that we cover the distribution of the characters w.r.t their positions
#count number of word-positions each character appears in
```

```python
cp_word_count = {(c,p): 0 for c in list('abcdefghijklmnopqrstuvwxyz')
                          for p in range(5)}

for word in solution_words:
    for p,c in enumerate(word):
        cp_word_count[(c,p)] += 1
```

```python
cp_word_count
```

```python
selected_cps = []

for i in range(15):
    
    for _ in selected_cps:
        cp_word_count[_] = 0

    max_cp_word_count = 0
    argmax_cp_word_count = None

    for word in useable_words:
        count = 0
        for p,c in enumerate(word):
            count += cp_word_count[(c,p)]
            
        if count>=max_cp_word_count:
            max_cp_word_count = count
            argmax_cp_word_count = word

    print('round: {}'.format(i))

    print('{}, {}\n'.format(argmax_cp_word_count, max_cp_word_count))
    
    selected_cps += [(c,p) for c in argmax_cp_word_count
                           for p in range(5)]
```
